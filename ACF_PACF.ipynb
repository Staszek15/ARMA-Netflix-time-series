```python
import numpy as np
import statsmodels.graphics.tsaplots as smgt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def acvf_empir(x, h):
    h = abs(h)
    mean_x = np.mean(x)
    n = len(x)
    
    from_h = x[h:]
    to_nh = x[:n-h]
    
    output = (1/n)*sum((from_h - mean_x)*(to_nh - mean_x))
    return output

def acf_empir(x, h):
    return acvf_empir(x,h)/acvf_empir(x,0)
```

```python
data = pd.read_csv("C:/Users/wypyc/Desktop/repozytoria/ARMA/ARMA-Netflix-time-series/data/general_data.csv")
data.rename(columns= {"Unnamed: 0" : "Date"}, inplace=True)
data.set_index("Date")
open_price = data.Open.to_list()
```

```python
plt.plot(open_price)
```

```python
H = np.arange(0,100)
ACF1 = [acf_empir(open_price, h) for h in H]
```

```python
plt.figure(figsize = (9,6))
plt.scatter(H,ACF1, color = "pink")
plt.title("Funkcja autokorelacji")
plt.xlabel("h")
```
